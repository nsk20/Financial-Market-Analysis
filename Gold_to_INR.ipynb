## Convert Price columns to numeric

### Subtask:
Convert the 'Price' column in `nifty_df` and `usd_inr_df` from string to numeric (float) by removing commas.


**Reasoning**:
To convert the 'Price' column in both `nifty_df` and `usd_inr_df` to a numeric type, I need to remove commas from the string values and then cast them to float.



```python
nifty_df['Price'] = nifty_df['Price'].astype(str).str.replace(',', '', regex=False).astype(float)
usd_inr_df['Price'] = usd_inr_df['Price'].astype(float)

print("Converted 'Price' column in nifty_df to numeric:")
print(nifty_df['Price'].head())
print("\nConverted 'Price' column in usd_inr_df to numeric:")
print(usd_inr_df['Price'].head())
```

## Prepare USD/INR for merge and interpolate

### Subtask:
Set the 'Date' column of `usd_inr_df` as the index, sort it, and then reindex it to cover all dates within its range. Interpolate any missing 'Price' values using linear interpolation to approximate rates for dates without direct matches.


**Reasoning**:
To prepare `usd_inr_df` for merging, I will set the 'Date' column as the index, sort it, reindex to a full date range, and then linearly interpolate any missing 'Price' values.



```python
usd_inr_df = usd_inr_df.set_index('Date')
usd_inr_df = usd_inr_df.sort_index()

date_range = pd.date_range(start=usd_inr_df.index.min(), end=usd_inr_df.index.max())
usd_inr_df = usd_inr_df.reindex(date_range)
usd_inr_df['Price'] = usd_inr_df['Price'].interpolate(method='linear')

print("Prepared USD/INR DataFrame:")
print(usd_inr_df.head())
print(usd_inr_df.tail())
```

**Reasoning**:
Now that `usd_inr_df` is prepared with a complete date range and interpolated 'Price' values, I will merge it with `nifty_df` on the 'Date' column. After the merge, I will calculate the `price_in_usd` column in `nifty_df` by dividing the 'Price' from `nifty_df` by the corresponding 'Price' (exchange rate) from the merged `usd_inr_df`.



```python
nifty_df = nifty_df.set_index('Date')
nifty_df = nifty_df.sort_index()

nifty_df = nifty_df.merge(usd_inr_df[['Price']], left_index=True, right_index=True, how='left', suffixes=('_nifty', '_usd_inr'))
nifty_df['price_in_usd'] = nifty_df['Price_nifty'] / nifty_df['Price_usd_inr']
nifty_df = nifty_df.reset_index()

print("Merged nifty_df with USD/INR data and calculated price_in_usd:")
print(nifty_df.head())
```

Summarize the calculated historical returns for each ticker ('Nifty-50', 'S&P500', 'Gold') over 1-year, 3-year, 5-year, 10-year, and 15-year periods. The returns should be based on the 'Price' column in the `combined_df` and presented in a new DataFrame. Before calculating returns, ensure that the 'Open', 'High', and 'Low' columns of the `Nifty-50` data in `combined_df` are converted to a numeric type, handling any commas present in the string values.

## Calculate Returns for Each Ticker

### Subtask:
Calculate historical returns for each ticker ('Nifty-50', 'S&P500', 'Gold') over 1-year, 3-year, 5-year, 10-year, and 15-year periods using the 'Price' column in `combined_df`, presenting the results in a new DataFrame. Ensure 'Open', 'High', and 'Low' columns for Nifty-50 data are numeric first.


```python
nifty_rows = combined_df['Ticker'] == 'Nifty-50'
for col in ['Open', 'High', 'Low']:
    combined_df.loc[nifty_rows, col] = combined_df.loc[nifty_rows, col].astype(str).str.replace(',', '', regex=False).astype(float)

combined_df['Date'] = pd.to_datetime(combined_df['Date'])
combined_df = combined_df.sort_values(by=['Ticker', 'Date']).reset_index(drop=True)

print("Converted 'Open', 'High', 'Low' for Nifty-50 to numeric and ensured Date column is datetime, then sorted the DataFrame:")
print(combined_df[combined_df['Ticker'] == 'Nifty-50'].head())
```

```python
returns_data = []
time_periods = [1, 3, 5, 10, 15, 20]

for ticker_name, group in combined_df.groupby('Ticker'):
    group = group.sort_values(by='Date')
    latest_date = group['Date'].max()
    latest_price = group.loc[group['Date'] == latest_date, 'Price'].iloc[0]

    for years in time_periods:
        target_start_date = latest_date - pd.DateOffset(years=years)
        # Find the earliest date in the group that is >= target_start_date
        start_period_data = group[group['Date'] >= target_start_date]

        if not start_period_data.empty:
            # Get the price corresponding to the earliest date in the filtered data
            start_period_price = start_period_data.sort_values(by='Date')['Price'].iloc[0]
            if start_period_price != 0: # Avoid division by zero
                returns = ((latest_price - start_period_price) / start_period_price) * 100
            else:
                returns = None # Handle cases where start_period_price is zero
        else:
            returns = None # Not enough data for the period

        returns_data.append({
            'Ticker': ticker_name,
            'Period': f"{years}-Year",
            'Return (%)': returns
        })

historical_returns_df = pd.DataFrame(returns_data)

print("Historical Returns for Each Ticker:")
historical_returns_df.head(len(time_periods) * len(combined_df['Ticker'].unique()))
```

# Task
Prepare `gold_df` and `usd_inr_df` for correlation analysis by ensuring their 'Date' columns are set as the index, sorting both DataFrames by date, and selecting only the 'Price' column for each.

## Prepare Data for Correlation Analysis

### Subtask:
Ensure the 'Date' columns in `gold_df` and `usd_inr_df` are properly formatted and set as index, and select the relevant 'Price' columns for analysis. Sort both dataframes by date.


**Reasoning**:
To prepare `gold_df` for correlation analysis, I will set the 'Date' column as its index, sort the index, and then select only the 'Price' column. For `usd_inr_df`, I will select only its 'Price' column, as it has already been processed with the 'Date' column as its index and sorted.



```python
gold_df = gold_df.set_index('Date').sort_index()[['Price']]
usd_inr_df = usd_inr_df[['Price']]

print("Prepared gold_df:")
print(gold_df.head())
print("\nPrepared usd_inr_df:")
print(usd_inr_df.head())
```

## Merge Gold and USD/INR Data

### Subtask:
Merge `gold_df` and `usd_inr_df` on their 'Date' column to create a combined DataFrame for correlation analysis. This will align the gold prices with the corresponding USD/INR exchange rates.


**Reasoning**:
To merge `gold_df` and `usd_inr_df` for correlation analysis, I will perform an inner join on their 'Date' index, rename the 'Price' columns for clarity, and then display the head of the resulting DataFrame.



```python
merged_df = gold_df.merge(usd_inr_df, left_index=True, right_index=True, how='inner', suffixes=('_gold', '_usd_inr'))
merged_df = merged_df.rename(columns={'Price_gold': 'Gold Price', 'Price_usd_inr': 'USD/INR Price'})

print("Merged Gold and USD/INR Data for correlation analysis:")
print(merged_df.head())
```

## Calculate Daily Percentage Change

### Subtask:
Calculate the daily percentage change for both gold prices and USD/INR exchange rates. This will help identify periods of price movement and INR depreciation.


**Reasoning**:
To calculate the daily percentage change for 'Gold Price' and 'USD/INR Price', I will use the `pct_change()` method on these columns in `merged_df` and multiply by 100 to express them as percentages.



```python
merged_df['Gold Daily Change (%)'] = merged_df['Gold Price'].pct_change() * 100
merged_df['USD/INR Daily Change (%)'] = merged_df['USD/INR Price'].pct_change() * 100

print("Merged DataFrame with daily percentage changes:")
print(merged_df.head())
```

## Calculate and Display Correlation

### Subtask:
Compute the correlation coefficient between the daily percentage change of gold prices and the daily percentage change of USD/INR. Display the correlation matrix or coefficient.


**Reasoning**:
To compute the correlation coefficient, I will select the 'Gold Daily Change (%)' and 'USD/INR Daily Change (%)' columns from `merged_df` and apply the `.corr()` method.



```python
correlation = merged_df['Gold Daily Change (%)'].corr(merged_df['USD/INR Daily Change (%)'])

print(f"Correlation between Gold Daily Change and USD/INR Daily Change: {correlation:.4f}")
```
